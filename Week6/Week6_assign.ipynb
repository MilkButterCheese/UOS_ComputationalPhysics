```python
import numpy as np
from scipy.interpolate import* 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#Problem1 
#Data 받아오고 설정하기
data= np.genfromtxt('/content/drive/MyDrive/ComputatinalPhysics/Week6/Data1.dat',delimiter=',')
t=data[0,:]
x=data[1,:]
y=data[2,:]
h=y[0]
#t-x data에서 v_0*cos(theta)값 구하기
def f(t,vcos): 
    return t*vcos  
#데이터에 적절한 Parameter 찾기
parm, var= curve_fit(f,t ,x)
# #Plot 그리기
tt=np.linspace(0,3.6,36)
xx=f(tt,parm)
plt.plot(t,x,'*')
plt.plot(tt,xx)
plt.plot()
print(parm)#v_0*cos(theta)=2.60269762
vcos=parm 
#t-y data에서 g,v_o*sin(theta)값 구하기
def f(t,g,vsin): 
    return -g*(t**2)/2 +vsin*t +h
#데이터에 적절한 Parameter 찾기
parm, var= curve_fit(f,t ,y)
#Plot 그리기
tt=np.linspace(0,3.6,36)
yy=f(tt,parm[0],parm[1])
print(yy)
plt.plot(t,y,'*')
plt.plot(tt,yy)
plt.plot()
print(parm) #g=9.78205393 v_0*sin(theta)=14.72799242
vsin=parm[1]
print(vsin/vcos)
print(np.sqrt(vsin**2+vcos**2))
#종합하여 h= 10.0578 g= g=9.78205393 tan(theta)=5.65874127, v_0= 14.9561959
```

```python
import numpy as np
from scipy.interpolate import* 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#Problem4
#Data 받아오고 설정하기
data= np.genfromtxt('/content/drive/MyDrive/ComputatinalPhysics/Week6/Data.dat',delimiter=',')
t=data[0,:]
x=data[1,:]

#함수의 모양이 정해져 있으므로
def f(t,A,r,w,phi): 
    return A*np.exp(-r*t)*np.sin(w*t+phi)
parm, var= curve_fit(f,t ,x)

# #Plot 그리기
tt=np.linspace(0,15,150)
xx=f(tt,parm[0],parm[1],parm[2],parm[3])
plt.plot(t,x,'*')
plt.plot(tt,xx)
plt.plot()
print(parm)
#A=0.48690504 b/2m=0.18035854 w=2.23097294 phi=1.56609573
#b=0.072143416 k=m*(w^2+(b/2m)^2)=1.0019538923926352
print(0.2*(2.23097294**2+0.18035854**2))
```
