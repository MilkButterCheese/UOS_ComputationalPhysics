```python
import numpy as np
import matplotlib.pyplot as plt
#Problem1 (가)
x=np.array([-1,1,3])
y=np.array([17,-7,-15])
n=len(x)
xx=np.linspace(x[0],x[-1],n*10)
m=len(xx)
L=np.zeros((m,n))

for j in range(0,m):
    L[j,0]= (xx[j]-x[1])*(xx[j]-x[2])/((x[0]-x[1])*(x[0]-x[2]))
    L[j,1]= (xx[j]-x[0])*(xx[j]-x[2])/((x[1]-x[0])*(x[1]-x[2]))
    L[j,2]= (xx[j]-x[0])*(xx[j]-x[1])/((x[2]-x[0])*(x[2]-x[1]))

for k in range(0,n):
    P= P+ y[k]*L[:,k]
#플롯 그리기
plt.plot(x,y,'*')
plt.plot(xx,P)
plt.plot
```
